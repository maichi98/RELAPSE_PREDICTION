```python
from pathlib import Path

import pandas as pd
import pydicom
```

```python
dir_patient = Path("/run/user/1001/gvfs/smb-share:server=31.10.11.4,share=tcga_images/data/col/AtezoLACC/atezolacc_data_cleaning/01-IGR/001-ST-C/curie_1.2.826.0.1.3680043.10.436.2645642819026852393015083111998964768/MR_20220214_1.2.826.0.1.3680043.10.436.8482883847529525727227509820639650874")

assert dir_patient.exists()
```

```python
def extract_metadata(path_slice):

    ds = pydicom.dcmread(path_slice)

    return {
        "PatientID": ds.PatientID,
        "StudyInstanceUID": ds.StudyInstanceUID,
        "SeriesInstanceUID": ds.SeriesInstanceUID,
        "SOPInstanceUID": ds.SOPInstanceUID,
        "Modality": ds.Modality,
        "SliceLocation": ds.SliceLocation,
        "InstanceNumber": ds.InstanceNumber,

    }
```

```python
df_data = pd.DataFrame(columns=["PatientID", "StudyInstanceUID", "SeriesInstanceUID", "SOPInstanceUID", "Modality", "SliceLocation", "InstanceNumber"])
slices = dir_patient.glob("*.dcm")

for path_slice in slices:

    df_data.loc[len(df_data)] = extract_metadata(path_slice)
```
