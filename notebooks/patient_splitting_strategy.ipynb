```python
from relapse_prediction import constants

import pandas as pd
import numpy as np
```

```python
df_data1 = pd.read_excel("splitting_strategy.xlsx", sheet_name="IDH et New KIKI Class")
df_data2 = pd.read_excel("splitting_strategy.xlsx", sheet_name="Sheet1")
```

```python
df_data1.rename(columns={"New_ID_AIDREAM": "patient", "IDH_mutated(IDHwt=0,IDHm=1)": "IDH_status", "new class(Biospy + STR <30%=1,STR >30% + GTR with L3=2,GTR without L3=3)": "Class"}, inplace=True)
```

```python
df_data2.rename(columns={"New_ID_AIDREAM": "patient", "TYPE_surgery(Bx=0,STR=1,GTR=2)": "surgery_type"}, inplace=True)
```

```python
df_data1 = df_data1[['patient', 'IDH_status', 'Class']].loc[df_data1["patient"].isin(constants.list_patients)]
df_data2 = df_data2[['patient', 'surgery_type']].loc[df_data2["patient"].isin(constants.list_patients)]
```

```python
df_splitting_strategy = pd.merge(df_data1, df_data2, on='patient', how='left')
```

```python
df_splitting_strategy.to_csv("splitting_strategy.csv", index=False)
```
