```python
from relapse_prediction import constants, utils

import pandas as pd
```

```python
df_ref = utils.get_referential_table()
```

```python
df_strat = pd.read_excel(constants.PATH_SPLITTING_STRATEGY)
```

```python
df_strat.rename(columns={"New_ID_AIDREAM": "AIDREAM_ID",
                         "IDH_mutated(IDHwt=0,IDHm=1)": "Class",
                         "new class(Biospy + STR <30%=1,STR >30% + GTR with L3=2,GTR without L3=3)": "Class"},
                inplace=True)
```

```python
df_ref = df_ref.merge(df_strat, on="AIDREAM_ID", how="left")
```

```python
df_ref.to_csv(constants.PATH_REFERENTIAL_TABLE, index=False)
```

```python
list_patients = df_ref["AIDREAM_ID"].tolist()
print(fr"Number of patients: {len(list_patients)}")
```

```python
constants.DIR
```

```python
df_syn = pd.DataFrame(columns=["AIDREAM_ID", "has SyN"])

for patient in list_patients:

    has_syn = True

    for stage in ["pre_RT", "Rechute"]:

        dir_syn = constants.DIR_PROCESSED / patient / "LABELS" / stage / "SyN"

        if not dir_syn.exists():
            has_syn = False

    df_syn.loc[len(df_syn)] = [patient, has_syn]

```

```python
df_ref = df_ref.merge(df_syn, on="AIDREAM_ID", how="left")
```
